```python
# from fiteanalytics.finx.client import FinXClient
from finx.client import FinXClient
import pandas as pd
import numpy as np
pd.options.plotting.backend = "plotly"
```

```python
fincad_api_key = '46c85416-f9c3-4bf9-aa1a-a0e5788dfca6'
finx_api_endpoint = 'http://52.34.48.102/api/' # FITE ANALYTICS DEVELOPMENT SERVER
finx = FinXClient('socket', finx_api_key=fincad_api_key, finx_api_endpoint=finx_api_endpoint)
```

```python
df = pd.DataFrame(finx._dispatch_batch(
    'var', 
    horizon_params=dict(years=1),
    input_file='./VaRPortfolio.csv',
    block=True))
df['total_market_value'] = df['par_value'].astype(float) * df['market_value'].astype(float)
df['VaR'] = df['VaR'].astype(float).map(lambda x: -1 * x if x > 0 else x)
df[['security_id', 'option_side', 'par_value', 'market_value', 'total_market_value', 'VaR']].rename(
    columns=dict(market_value='price', total_market_value='market_value')).to_csv('VaROutput_1Yr.csv', index=False)
```
